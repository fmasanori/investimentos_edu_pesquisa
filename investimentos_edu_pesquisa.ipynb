```python
df_2010 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2010.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2011 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2011.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2012 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2012.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2013 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2013.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2014 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2014.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2015 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2015.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2016 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2016.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DES_PESQUISA'])
df_2017 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2017.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DESPESA_PESQUISA'])
df_2018 = pd.read_csv(r'C:\Users\Agnes\Desktop\EP1\DM_IES_2018.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['VL_DESPESA_PESQUISA'])
```

```python
factor = 10**(-9)
data_2010 = (df_2010.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2011 = (df_2011.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2012 = (df_2012.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2013 = (df_2013.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2014 = (df_2014.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2015 = (df_2015.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2016 = (df_2016.query('VL_DES_PESQUISA != 0')['VL_DES_PESQUISA'].sum())*factor
data_2017 = (df_2017.query('VL_DESPESA_PESQUISA != 0')['VL_DESPESA_PESQUISA'].sum())*factor
data_2018 = (df_2018.query('VL_DESPESA_PESQUISA != 0')['VL_DESPESA_PESQUISA'].sum())*factor
```

```python
left = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
height = [data_2010, data_2011, data_2012, data_2013, data_2014, data_2015, data_2016, data_2017, data_2018]
tick_label = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']
plt.rcParams['figure.figsize'] = (15, 10)
plt.bar(left, height, tick_label = tick_label, width = 0.4, color = ['indianred'])
plt.xlabel('Anos')
plt.ylabel('Total investido (em bilh√µes de R$)')
plt.title('Valores investidos em Pesquisa e Desenvolvimento de 2010 a 2018')
for i, v in enumerate(height): # Para exibir os valores de cada conjunto de dados acima das barras 
    v = round(v, 2) #Para arredondar os valores
    plt.text(left[i] - 0.13, v + 0.03, str(v))
plt.show()
```
